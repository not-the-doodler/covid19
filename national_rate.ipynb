```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('https://api.covid19india.org/csv/latest/case_time_series.csv')
df = pd.DataFrame(dataset)
tc = df["Total Confirmed"].to_numpy()
td = df["Total Deceased"].to_numpy()
tr = df["Total Recovered"].to_numpy()
dc = df["Daily Confirmed"].to_numpy()
dd = df["Daily Deceased"].to_numpy()
```

```python
growthrate=[]
for i in range (0,len(tc)):
    growthrate.append((dc[i]/tc[i])*100)
avg_gr=np.mean(growthrate)  
avg_gr
plt.plot (growthrate)
```

```python
deathrate_normal=[]
for i in range (1,len(td)):
    deathrate_normal.append((td[i]/tc[i-7])*100)
avg_drn=np.mean(deathrate_normal) 
plt.plot(deathrate_normal)
avg_drn
```

```python
deathrate=[] # (deaths at day.x / cases at day.x-{T}) (T=7)
for i in range (7,len(dd)):
    if dc[i-7]==0:
        continue
    deathrate.append((dd[i]/dc[i-7])*100)
avg_dr=np.mean(deathrate) 
plt.plot(deathrate)
```

```python
recoveryrate=[]
for i in range (1,len(td)):
    recoveryrate.append((tr[i]/tc[i-1])*100)
avg_rr=np.mean(recoveryrate) 
avg_rr
plt.plot (recoveryrate)
```
