```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('https://api.covid19india.org/csv/latest/statewise_tested_numbers_data.csv')
df = pd.DataFrame(dataset)
states=['Maharashtra', 'Kerala', 'West Bengal', 'Odisha', 'Bihar']
states=['Maharashtra','Gujarat', 'Delhi', 'Rajasthan', 'Madhya Pradesh', 'Tamil Nadu', 'Uttar Pradesh', 'Andhra Pradesh']
```

```python
plt.figure()
ax = plt.subplot(111)
for i in states:
    ap=df[df['State'].str.contains(i)]
    test=ap["Total Tested"].to_numpy()
   # test = test[~np.isnan(test)]
    positive=ap["Positive"].to_numpy()
    #positive = positive[~np.isnan(positive)]
    ax.plot(test,positive, linewidth=1)
    ax.scatter(test,positive,s=10)
ax.set_title("Total positives vs Total Tests")
ax.legend(states)
ax.set_xlabel('Total Tests')
ax.set_ylabel('Total Positives')
plt.savefig('test.png',dpi=1000,bbox_inches='tight')
```
